### Count the Different States and plot all in ordering

```python
df['State'].value_counts()
```

```python
plt.figure(figsize=(15,6))
sns.countplot(data=df,x='State',order=df['State'].value_counts().index,palette='rocket_r')
plt.xticks(rotation=90)
plt.show()
```

#### Highest number of buyers are from California, New York and Texas.

### Profit vs Discount

```python
fig,ax=plt.subplots(figsize=(12,6))
ax.scatter(df['Sales'],df['Profit'],color='brown')
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Discount',y='Profit',label='Profit',data=df,color='brown')
plt.legend()
plt.show()
```

#### From above trend line Graph we can say that Profit and Discount are negatively correlated to eachothers.

### Profit vs Quantity

```python
plt.figure(figsize=(12,6))
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=df,color='brown')
plt.legend()
plt.show()
```

```python
df.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['brown','red'],figsize=(12,6))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

#### From the above hisogram we can see Profit and sales are maximum in consumer segment and minimum in Home Office segment

```python
df.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['brown','red'],figsize=(12,6))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

#### Observation : Profit and sales are maximum in west region and minimum in south region

```python
t_states = df['State'].value_counts().nlargest(10)
t_states
```

```python
df.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['brown','red'],alpha=0.9,figsize=(12,6))
plt.ylabel('Profit/Loss and sales')
plt.show()
```

### As a business manager, try to find out the weak areas where you can work to make more profit?

   (I) Technology and Office Supplies have high profit.
    
   (II) Furniture have less profit.

```python
ps = df.groupby('Sub-Category')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['red','brown'],figsize=(15,8))
plt.title('Profit/loss & Sales across states')
plt.xlabel('Sub-Category')
plt.ylabel('Profit/loss & Sales')
plt.show()
```

### Observation:
    (I)Phones sub-category have high sales. 
    
    (II)chairs have high sales but less profit compared to phones.
    
    (III)Tables and Bookmarks sub-categories facing huge loss.
